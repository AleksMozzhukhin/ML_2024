```python
df=table
df.publishedAt = pd.to_datetime(df.publishedAt)
```

```python
df.publishedAt.dt.date
```

```python
import matplotlib.pyplot as plt #     

data_without_repetitions = df.drop_duplicates(subset=['video_id'], keep='first')
dates_count = data_without_repetitions['publishedAt'].dt.date.value_counts().sort_index()
dates_for_plot = pd.to_datetime(pd.date_range(start=dates_count.index.min(),
                                              end=dates_count.index.max(), freq="D").to_list()).date
dates_counts_for_plot = []
for date in dates_for_plot:
    dates_counts_for_plot.append(dates_count[date])
fig, ax = plt.subplots()
ax.xaxis_date()
fig.autofmt_xdate()
plt.plot(dates_for_plot, dates_counts_for_plot)
plt.xlabel(' ')
plt.ylabel('  ')
plt.title('     ')
plt.grid()
plt.show()


```

### <font color='DarkOrange'>** 2 [0.5 a]**</font>

  `(bar plot)` -      .

```python
plt.figure(figsize=(12, 6))
df['country'].value_counts().plot(kind='bar', color='skyblue')
plt.title('    ')
plt.xlabel('')
plt.ylabel('  ')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### <font color='DarkOrange'>** 3 [1 ]**</font>

  `(scatter plot)`  -     -      (country='FR').          (comments_disabled=True).

    ?

```python
france_videos = df[(df['country'] == 'FR') & (df['comments_disabled'] == False)]
plt.figure(figsize=(10, 6))
plt.scatter(france_videos['view_count'], france_videos['comment_count'], alpha=0.6, color='blue')
plt.title('      ()')
plt.xlabel(' ')
plt.ylabel(' ')
plt.xscale('log')  #      x
plt.yscale('log')  #      y
plt.grid(True)
plt.tight_layout()
plt.show()
```
